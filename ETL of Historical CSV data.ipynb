Overview of datasets

These datasets are historical price data of Gold (per troy ounce), Crude Oil (per barrel), Bitcoin (per coin), and SNP500 (per stock).
Source: https://www.investing.com/
Timeframe: 1/1/2015 - 28/11/2022

Each datasets contains: Date, average price, open, high (max price in the day), low (max price in the day), trading volume, change(%).

Aim: Join all high-low price of all commodities/cryptocurrency/stock into one dataframe.

```python
import pandas as pd
```

```python
df_bitcoin_price = pd.read_csv("source/Bitcoin Historical Data.csv")
df_oil_price = pd.read_csv("source/Crude Oil WTI Futures Historical Data.csv")
df_gold_price = pd.read_csv("source/Gold Futures Historical Data.csv")
df_snp500_price = pd.read_csv("source/Crude Oil WTI Futures Historical Data.csv")
df_usd_euro_rate = pd.read_csv("source/EUR_USD Historical Data.csv")
```

```python
df_bitcoin_price = df_bitcoin_price.drop(columns=["Price", "Open", "Vol.", "Change %"])
df_oil_price = df_oil_price.drop(columns=["Price", "Open", "Vol.", "Change %"])
df_gold_price = df_gold_price.drop(columns=["Price", "Open", "Vol.", "Change %"])
df_snp500_price = df_snp500_price.drop(columns=["Price", "Open", "Vol.", "Change %"])
df_usd_euro_rate = df_usd_euro_rate.drop(columns=["Open", "High", "Low", "Vol.", "Change %"])
```

```python
#due to differences in the datetime format in bitcoin with other dfs, i replace it with snp500's date column
df_bitcoin_price["Date"] = df_snp500_price["Date"].astype(str)
df_oil_price["Date"] = df_oil_price["Date"].astype(str)
df_gold_price["Date"] = df_gold_price["Date"].astype(str)
df_snp500_price["Date"] = df_snp500_price["Date"].astype(str)
df_usd_euro_rate["Date"] = df_usd_euro_rate["Date"].astype(str)
```

```python
df_bitcoin_price = df_bitcoin_price.rename(columns={"High":"Bitcoin_High", "Low":"Bitcoin_Low"})
df_oil_price = df_oil_price.rename(columns={"High":"Oil_High", "Low":"Oil_Low"})
df_gold_price = df_gold_price.rename(columns={"High":"Gold_High", "Low":"Gold_Low"})
df_snp500_price = df_snp500_price.rename(columns={"High":"SNP500_High", "Low":"SNP500_Low"})
df_usd_euro_rate = df_usd_euro_rate.rename(columns={"Interest":"US_Interest_Rate","Price":"EURO_To_USD_Rate"})
```

```python
df_snp500_price = df_snp500_price.merge(df_bitcoin_price, left_on='Date', right_on='Date')
df_snp500_price = df_snp500_price.merge(df_oil_price, left_on='Date', right_on='Date')
df_snp500_price = df_snp500_price.merge(df_gold_price, left_on='Date', right_on='Date')
df_snp500_price = df_snp500_price.merge(df_usd_euro_rate, left_on='Date', right_on='Date')
df_snp500_price["Interest"] = 0
df_snp500_price
```

```python
Interest_Year = {2022:1.41,2021:0.08,2020:0.36,2019:2.16,2018:1.79, 2017:1.00, 2016:0.39, 2015:0.13}
df_snp500_price["Year"] = pd.DatetimeIndex(df_snp500_price['Date']).year
df_snp500_price["Interest"] = 0
```

```python
for i in range(len(df_snp500_price)):
    df_snp500_price["Interest"][i] = Interest_Year[df_snp500_price["Year"][i]]
```
